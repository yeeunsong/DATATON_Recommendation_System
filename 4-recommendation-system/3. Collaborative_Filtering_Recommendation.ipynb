## 유저 시뮬레이션
키워드 하나 뽑아서 해당되는 동아리 중 랜덤으로 뽑는다

```python
def simulation(keyword, userId):
    index = tags[tags.apply(lambda x: True if keyword in x else False)].index
    
    #index중에서 len(index)//2+1개만큼 픽 
    select = np.random.choice(index,len(index)//2+1, replace=False)
    select = pd.Series(select,name='clubId')
    
    #select횟수만큼 userid출력
    userids = pd.Series(np.repeat(userId,len(select)),name='userId')
    
    #concat
    dd=pd.concat([userids, select], axis=1)
    dd['viewed']=1
    return dd
```

```python
simulation('봉사',1)
```

```python
alll = []
for t in tags:
    alll+=t
```

```python
alll[:10]
```

```python
np.random.choice(alll)
```

```python
usertable = pd.DataFrame(columns=['userId','clubId','viewed'])
for _ in range(100):
    kw = np.random.choice(alll)
    usertable = pd.concat([usertable, simulation(kw, _+1)])
```

# Latent Factor Collaborative Filtering (잠재요인 협업 필터링)
using matrix factorization

```python
user_item = item_user.T
```

```python
user_item.columns = clubnames
user_item
```

```python
from numpy.linalg import svd

U, sigma, Vt = svd(user_item)
```

```python
len(sigma)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.lineplot(x = range(len(sigma)), y = sigma)
plt.show()
```

```python
#find cumsum
cumratio = list(map(lambda x : round(x,3), np.cumsum(sigma)/np.sum(sigma)))
print(cumratio)
```

```python
print(len(sigma))
print(len(list(filter(lambda x : x<0.8, cumratio))))
#80%의 설명력을 가지고 있는 요소들만 pick => 약 1/2부분
```

```python
from scipy.sparse.linalg import svds #does truncated svd

U, sigma, Vt = svds(user_item, k = 48) #k=남길 특이값 개수
```

```python
print(U.shape, sigma.shape, Vt.shape)
```

```python
sigma[-10:]
```

```python
#sigma is given as vector => convert to matrix
sigma = np.diag(sigma)
```

```python
#convert back to original matrix by matrix multiplaction
svd_pred = np.dot(np.dot(U,sigma),Vt)
svd_pred = pd.DataFrame(svd_pred)
svd_pred #값이 달라짐 - truncated를 썼기 때문
```

```python
#sort expected ratings
exp_ratings = svd_pred.loc[10].sort_values(ascending=False)
exp_ratings
```

```python
#sort expected ratings get index
sorted_index = svd_pred.loc[50].sort_values(ascending=False).index
sorted_index
```

```python
#조회했던 동아리들
seen_clubs = svd_pred.columns[user_item.loc[50]!=0]
seen_clubs
```

```python
seen_names = user_item.loc[50][user_item.loc[50]!=0].index
list(seen_names.unique())
```

```python
#안본것들 중에서 가장 취향일만한 동아리 다섯가지
sorted_notseen = [x for x in sorted_index if x not in seen_clubs]
user_item.columns[sorted_notseen][:5]
```

```python
def svd_recommend(user_id, num_recom=5):
    index = user_id-1
    
    #sort expected ratings
    sorted_index = svd_pred.loc[user_id].sort_values(ascending=False).index
    
    #get list of clubs seen by user
    seen_names = user_item.loc[user_id][user_item.loc[user_id]!=0].index
    seen_names = list(seen_names.unique())

    #get top n from movies user has not seen
    sorted_notseen = [x for x in sorted_index if x not in seen_clubs]
    recom_title = list(user_item.columns[sorted_notseen][:num_recom])
    
    print('what you have seen : {}'.format(seen_names))
    print('Recommended for you : {}'.format(recom_title))
```

```python
svd_recommend(50)
```
