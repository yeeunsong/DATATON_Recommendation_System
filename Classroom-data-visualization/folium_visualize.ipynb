## Load Data - Buildings

```python
f1 = open("most_visited_building.txt", "r", encoding='UTF-8-sig')
Lines = f1.readlines()
```

```python
# check if the loaded data is proper
i = 0
for line in Lines:
    #print(line)
    i+=1
#print(i)
```

```python
building_list = []
numbers_list = []
for line in Lines:
    line = re.sub("\[\"","", line)
    line = re.sub("\"","", line)
    line = re.sub("\]\n","", line)
    split_word = line.split(", ")
    building_list.append(split_word[0])
    numbers_list.append(float(split_word[1]))
    
print(building_list)
print(numbers_list)
```

```python
df = pd.DataFrame({"building": building_list, "numbers": numbers_list, "place_list": place_list, "place_list_name": place_list_name})
```

```python
df
```

```python
# 의대 데이터 전처리
listofBuildings = ["우정간호학관", "의대제1강의실", "의대제2강의실", "의대제3강의실", "의대제4강의실", "의과대학본관"]
num_sum=0
df2 = df.iloc[33,:]
# print(append)

for index, row in df.iterrows():
    if row["building"] in listofBuildings:
        num = row["numbers"]
        num_sum+=num
        df.drop(index, inplace=True)

df = df.append(df2, ignore_index=True)
print(type(df))
is_doc = df.loc[df["building"]=="의과대학본관", "numbers"] = num_sum

df.reset_index(drop=True, inplace=True)
```

```python
df
f1.close()
```

## Visualize Data - Buildings

```python
map1 = folium.Map(location=latlong, zoom_start=20)
```

```python
for index, row in df.iterrows():
    folium.Circle(
        location= row["place_list"],
        radius= math.sqrt(row['numbers'])*0.2,
        color='crimson',
        popup = row["building"]+', '+str(int(row["numbers"])),
        fill=True,
        fill_color='crimson'
).add_to(map1)
```

## Load Data - Routes

```python
f1 = open("most_used_routes.txt", "r+", encoding='UTF-8-sig')
Lines = f1.readlines()
```

```python
for line in Lines:
    print(line)
    
f1.close()
```

```python
start_list = []
end_list = []
used_num_list = []
```

```python
for line in Lines:
    line = re.sub("\[\"","", line)
    line = re.sub("\"","", line)
    line = re.sub("\]\n","", line)
    
    split_word = line.split(", ")
    used_num_list.append(split_word[1])
   
    start_end = split_word[0].split(":")
    start_list.append(start_end[0])
    end_list.append(start_end[1])
    
print(start_list)
print(end_list)
print(used_num_list)
```

```python
print(len(start_list))
print(len(end_list))
print(len(used_num_list))
```

```python
# merge list to dataframe
df2 = pd.DataFrame({"start": start_list, "end": end_list, "used_num": used_num_list})
```

```python
df2
```

```python
# place_list: list of building coordinates
# place_list_name: name of buildings
# check the names of the buildings: 데이터 전처리하기

for index, row in df2.iterrows():
    if row["start"] not in place_list_name:
        print(row["start"])
    if row["end"] not in place_list_name:
        print(row["end"])
```

```python
# 의대 건물 이름 의과대학본관으로 모두 일치시키기

doc_list = ["의대", "우정간호학관", "의대제1강의실", "의대제2강의실", "의대제3강의실", "의대제4강의실"]

df2.replace(doc_list, "의과대학본관", inplace=True)
```

```python
# 의대 건물 이름 전처리 잘 되었는지 확인
for index, row in df2.iterrows():
    if row["start"] in doc_list:
        print(row["start"])
    if row["end"] in doc_list:
        print(row["end"])
```

## Visualize Data - Routes

```python
map2 = folium.Map(location=latlong, zoom_start=20)
```

```python
df2.head()
print(len(df2))
```

```python
# merge 2 datasets so coordinates can be shown (coordinates of start, coordinates of end building)
# merge the start building coordinate to df

df_for_merge  = df.copy()
df_for_merge["start"] = df_for_merge["building"]
df_for_merge.rename(columns={"place_list":"start_coordinate"}, inplace=True)
df_for_merge.drop(['numbers', 'place_list_name', "building"], axis=1, inplace=True)
df_for_merge.head()
df2 = df2.merge(df_for_merge, on="start", how='outer').fillna('')
```

```python
# merge 2 datasets so coordinates can be shown (coordinates of start, coordinates of end building)
# merge the end building coordinate to df

df_for_merge  = df.copy()
df_for_merge["end"] = df_for_merge["building"]
df_for_merge.rename(columns={"place_list":"end_coordinate"}, inplace=True)
df_for_merge.drop(['numbers', 'place_list_name', "building"], axis=1, inplace=True)
df_for_merge.head()
df2 = df2.merge(df_for_merge, on="end", how='outer').fillna('')
```

```python
coordinates = []

for index, row in df2.iterrows():
    coordinates.append([row["start_coordinate"], row["end_coordinate"]])
    
print(len(coordinates))
```

```python
df2["coordinates"] = coordinates
```

```python
df2.head()
```

```python
for index, row in df2.iterrows():
    folium.PolyLine(
        locations = row["coordinates"],
        weight = float(row["used_num"])*0.0006,
        color='crimson',
        popup = row["start"]+'to'+row["end"],

).add_to(map2)
```
