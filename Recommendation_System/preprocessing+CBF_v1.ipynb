## Text Preprocessing (Without KoNLPy)

```python
import pandas as pd
import re
clubs = pd.read_csv('clubs.csv')
clubs.head()
```

```python
clubs.info()
```

```python
clubs['txt'] = clubs['title']+clubs['description']
clubs.head()
```

```python
def cleantxt(txt):
    txt = txt[:-35]
    txt = txt.replace('\n',' ')
    txt = txt.replace('\r','')
    p = re.compile(r'[0-9A-Za-z가-힣?.!\'\(\) ]+')
    return ''.join(re.findall(p,txt))
```

```python
clubs['txt'] = clubs['txt'].apply(cleantxt)
```

```python
clubs.txt.head()
```

```python
stopwords = ['의', '가', '이', '은', '들', '는', '좀', '잘', '걍', '과', '도', '를', '으로', '자', '에', '와', '한', '하다']
```

## CBF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#ngram, stopwords 기능 지원
#객체 생성
tfidf = TfidfVectorizer(analyzer = 'word', ngram_range=(1,2), min_df=0, stop_words = stopwords)
```

```python
tfidf_matrix = tfidf.fit_transform(clubs.txt)
```

```python
tfidf_matrix.shape
```

```python
print(tfidf_matrix[10])
```

```python
from sklearn.metrics.pairwise import linear_kernel
#linear_kernel = dot product
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim[0] #1번동아리 - 다른 동아리 간 유사도
```

```python
titles = clubs['title']
indices = pd.Series(clubs.index, index=clubs['title'])

print(titles.head(), indices.head())
```

```python
list(enumerate(cosine_sim[0]))
```

```python
def get_recommendations(title):
    i = indices[title]
    sim_scores = list(enumerate(cosine_sim[i]))
    sim_scores = sorted(sim_scores, key = lambda x : x[1], reverse=True)
    sim_scores = sim_scores[1:21] 
    recommend = titles[[s[0] for s in sim_scores]]
    return recommend
```
