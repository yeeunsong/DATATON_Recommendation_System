```python
def avoid_null(data, header):
    data[header] = data[header].fillna('')
    return data[header]

def cos_sim(A, B):
    return dot(A, B) / (norm(A)*norm(B))  # 코사인 유사도 계산

def tfidf(df, TfidVectorizer):
    df['title'] = avoid_null(df, 'title')
    tfidf_metrix_of_tit = TfidVectorizer.fit_transform(df['title']) # 입력받은 제목의 값을 가져와 tfidf계산
    tfidf_metrix_of_des = TfidVectorizer.fit_transform(df['description'])
    
    return tfidf_metrix_of_tit, tfidf_metrix_of_des


# tf-idf : 여러 문서로 이루어진 문서군에서 어떤 단어가 특정 문서 내에서 얼마나 중요한 지를 나타내는 수치
# tf - term frequency (특정 단어가 문서내에서 나타나는 빈도수)
# df - document frequency (단어가 문서군에서 나타나는 빈도수)
```

```python
def top10_indices(data, q_num):
    cos_sim = linear_kernel(data, data)  # 입력 데이터의 코사인 유사도
    cos_sim_score = list(enumerate(cos_sim[q_num]))
    cos_sim_score = sorted(cos_sim_score, key=lambda x : x[1], reverse=True)
    score = cos_sim_score[1:11]
    tag_indices = [i[0] for i in score]
    
    return tag_indices

# 현재 보고 있는 게시글(q_num)과 가장 유사한 10개의 게시글 보기
```

```python
indices = pd.Series(df.index, index=df.title).drop_duplicates()
print(indices)

# 인덱스가 제목이고 번호가 값인 series 배열
# 해당 배열에 title명을 입력하면 그에 대응하는 인덱스 번호가 추출
```

```python
def rec(title, cosine_sim=cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx])) # 모든 입력에 대해서 해당 입력과의 유사도 구하기
    sim_scores = sorted(sim_scores, key=lambda x:x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    
    sim_indices = [i[0] for i in sim_scores]
    
    result_df = df.iloc[sim_indices].copy() 
    result_df['score'] = [i[1] for i in sim_scores]
    
    return result_df
    
```
