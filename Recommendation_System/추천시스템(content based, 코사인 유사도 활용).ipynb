```python
def avoid_null(data, header):
    data[header] = data[header].fillna('')
    return data[header]

def cos_sim(A, B):
    return dot(A, B) / (norm(A)*norm(B))  # ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê³„ì‚°

def tfidf(df, TfidVectorizer):
    df['title'] = avoid_null(df, 'title')
    tfidf_metrix_of_tit = TfidVectorizer.fit_transform(df['title']) # ì…ë ¥ë°›ì€ ì œëª©ì˜ ê°’ì„ ê°€ì ¸ì™€ tfidfê³„ì‚°
    tfidf_metrix_of_des = TfidVectorizer.fit_transform(df['description'])
    
    return tfidf_metrix_of_tit, tfidf_metrix_of_des


# tf-idf : ì—¬ëŸ¬ ë¬¸ì„œë¡œ ì´ë£¨ì–´ì§„ ë¬¸ì„œêµ°ì—ì„œ ì–´ë–¤ ë‹¨ì–´ê°€ íŠ¹ì • ë¬¸ì„œ ë‚´ì—ì„œ ì–¼ë§ˆë‚˜ ì¤‘ìš”í•œ ì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ìˆ˜ì¹˜
# tf - term frequency (íŠ¹ì • ë‹¨ì–´ê°€ ë¬¸ì„œë‚´ì—ì„œ ë‚˜íƒ€ë‚˜ëŠ” ë¹ˆë„ìˆ˜)
# df - document frequency (ë‹¨ì–´ê°€ ë¬¸ì„œêµ°ì—ì„œ ë‚˜íƒ€ë‚˜ëŠ” ë¹ˆë„ìˆ˜)
```

```python
def top10_indices(data, q_num):
    cos_sim = linear_kernel(data, data)  # ì…ë ¥ ë°ì´í„°ì˜ ì½”ì‚¬ì¸ ìœ ì‚¬ë„
    cos_sim_score = list(enumerate(cos_sim[q_num]))
    cos_sim_score = sorted(cos_sim_score, key=lambda x : x[1], reverse=True)
    score = cos_sim_score[1:11]
    tag_indices = [i[0] for i in score]
    
    return tag_indices

# í˜„ì¬ ë³´ê³  ìˆëŠ” ê²Œì‹œê¸€(q_num)ê³¼ ê°€ì¥ ìœ ì‚¬í•œ 10ê°œì˜ ê²Œì‹œê¸€ ë³´ê¸°
```
