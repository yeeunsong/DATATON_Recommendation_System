# '취업 프로그램 신청' 크롤링

```python
## '취업 프로그램 신청' 사이트 링크: https://job.korea.ac.kr/user/recu/program/pgmScheduleList.do
```

```python
urls = []
title = []
description = []
info = []
image = []
tags = []
```

```python

for page in range(300):
    url = "https://job.korea.ac.kr/user/recu/program/pgmScheduleView.do?gjobPgmNo="+str(525-page)
    urls.append(url)
```

```python

for i in urls:
    driver.get(i)
    raw = driver.page_source
    soup = BeautifulSoup(raw, 'html.parser')


    title.append(soup.select("td.board_view_title")[0].text)
    description.append(soup.select("div.board_view_content")[0].text)
    info.append(soup.find('tbody').text)

    if soup.find('img') != None:
        imgsrc = soup.find_all('img')[1].attrs['src']
    else:
        imgsrc = None

    image.append(imgsrc)
```

```python

df = pd.DataFrame(list(zip(urls, title, description, info, image)),
                  columns =['urls', 'title', 'description', 'info', 'image'])

df['tags'] = "취업"
```

```python
# 내려간 글도 있어서 title이 NAN이면 제거

indexNames = df[ df['title'] == '' ].index
df.drop(indexNames , inplace=True)
```

```python
# info column 전처리

df['info'] = df['info'].apply(lambda x: x.replace('\t', ''))
df['info'] = df['info'].apply(lambda x: x.replace('\n', ''))
```

```python
df = df.reset_index()
```
