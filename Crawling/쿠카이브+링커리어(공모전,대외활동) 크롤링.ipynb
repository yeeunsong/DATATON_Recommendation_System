## 셀레니움 사용

```python
from selenium import webdriver
import unicodedata
from datetime import timedelta, datetime
from requests import request
from requests.compat import urljoin, urlparse
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome('C:/Users/user/chromedriver_win32/chromedriver.exe')
```

### 공모전

```python
import sqlite3

conn = sqlite3.connect('linkcareer.db')
cur = conn.cursor()
```

```python
cur.executescript('''
    DROP TABLE IF EXISTS contest;
    CREATE TABLE contest(
        id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
        title TEXT NOT NULL,
        link TEXT NOT NULL,
        host TEXT,
        duedate TEXT,
        image TEXT,
        description TEXT,
        tags TEXT
    );
''')
```

```python
#조회순 top200
for i in range(1,11):
    print(i)
    url = 'https://linkareer.com/list/contest?filterType=CATEGORY&orderBy_direction=DESC&orderBy_field=VIEW_COUNT&page='+str(i)
    driver.get(url)
    dom = BeautifulSoup(driver.page_source, 'html.parser')
    
    node = dom.find_all('div',{'class':'jss396'})

    #for every activity
    for _ in node:
        #링크
        link = _.find('a').attrs['href']
        link = urljoin(url,link)
        
        #공모전제목
        title = _.find('h5').text
        
        #주최
        host = _.find('p').text
        
        #마감일(디데이)
        due = _.find('h4').text
        if due=='오늘마감':
            due = '08/23/2020'
        else:
            due = due[2:] #get only integer part
            due = datetime.now()+ timedelta(days=int(due))
            due = str(due.month)+'/'+str(due.day)+'/'+str(due.year) #get exact date
        
        #해당 공모전 페이지에 엑세스
        driver.get(link)
        dom = BeautifulSoup(driver.page_source, 'html.parser')
        
        node = dom.find('div',{'class':'ActivityDetailTab__ContentDetailContainer-jmpzn9-3 fgxpDE container'})
        
        #이미지
        if node.find('img')!=None:
            imgsrc = node.find('img').attrs['src']
        else:
            imgsrc = None
        
        #본문
        body = node.find('div').text
#         body = unicodedata.normalize("NFKD", body)
        
        #태그
        tags = []
        for _ in dom.find('div',{'class':'InformationBox__InterestField-n4qp1q-16 eOJAbb field is-horizontal'}).find_all('span'):
            tags.append(_.text)
        tags = str(tags)    
        

        print(title)
        cur.execute('INSERT INTO contest VALUES(NULL,?,?,?,?,?,?,?)',[title,link,host,due,imgsrc,body,tags])
        conn.commit()
```

```python
conn.close()
cur.close()
```

# 공모전/대외활동 데이터프레임 tag 수정

### 공모전

```python
df = pd.read_csv('contest.csv')
df.tags
```

```python
import ast

tags = df.tags.apply(ast.literal_eval)
```

```python
tmp = set()
for _ in tags:
    for i in _:
        tmp.add(i)
```

```python
list(tmp)
```

```python
replaceee = {'디자인/사진/예술/영상':'미술',
 '경제/금융':'경영',
 '유통/물류':'경영',
 '창업/자기계발':'경영',
 '문화/역사':'사회',
 '정치/사회/법률':'사회',
 '과학/공학/기술/IT':'IT',
 '콘텐츠':'경영',
 '경영/컨설팅/마케팅':'경영',
 '교육':'사회',
 '여행/호텔/항공':'사회',
 '언론/미디어':'경영',
 '사회공헌/교류':'봉사',
 '환경/에너지':'과학',
 '행사/페스티벌':'미술',
 '의료/보건':'과학',
 '공모전':'공모전'}
```

```python
for j in range(len(tags)):
    l = len(tags[j])
    for i in range(l):
        if tags[j][i] in replaceee.keys():
            tags[j][i] = replaceee[tags[j][i]]
        elif tags[j][i]=='과학/공학/기술/IT':
            tags[j][i]='IT'
            tags[j].append('과학')
        else:
            tags[j][i]='기타'
```

```python
for _ in tags:
    if '기타' in _:
        _.remove('기타')
```

```python
tags
```

```python
df.head()
```

```python
df.tags = tags
```

```python
df = df.iloc[:,1:]
df.head()
```

```python
df.to_csv('contest_tagfixed.csv')
```
